### Index vs. Method Chaining

**Index Chaining** is a bad practice:
```
df.loc[“Washtenaw”][“Total Population”]
```
You never know, when a view or a copy is returned and on what you work (depends on underlaying `numpy`):

-- <cite>If you see a ][ you should think carefully about what you are doing (Tom
Augspurger)</cite>

**Method Chaining** is best practice.

````python
# good code style:   m e t h o d s   in multiple lines
(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME','CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))

# makes a   c o p y !
````

# Group by

````python
import pandas as pd
import numpy as np
df = pd.read_csv('../../data/census.csv')
df = df[df['SUMLEV']==50]
df[:5]
````

````python
%%timeit -n 10
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in state ' + state + ' have an average population of ' + str(avg))
````

````python
%%timeit -n 10
for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + ' have an average population of ' + str(avg))
````

````python
df.head()
````

````python
df = df.set_index('STNAME')

def fun(state_name):
    if state_name[0]<'M':
        return 0
    if state_name[0]<'Q':
        return 1
    return 2

for group, frame in df.groupby(fun):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing.')
````

# Scales

````python
df = pd.DataFrame(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D'],
                  index=['excellent', 'excellent', 'excellent', 'good', 'good', 'good', 'ok', 'ok', 'ok', 'poor', 'poor'])
df.rename(columns={0: 'Grades'}, inplace=True)
df
````

````python
grades = df['Grades'].astype('category').head()
````

grades > 'C' # -> does not work

````python
grades = df['Grades'].astype('category',
                             categories=['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'],
                             ordered=True)
grades.head()
````
